## bar plot
* 범주향 데이터 요약 및 시각적 비교
* 각 범주의 값의 갯수 차이를 비교하고 갯수차이가 극단적인지 확인
* 차이가 극단적일 경우에는 전처리과정에서 업/다운 샘플링을 통해 각 범주의 개수가 유사하도록 조정
</br></br>
* data_cnt = data['target'].value_counts()
* plt.bar(x, heught, width = 0.8, bottom = None, align = 'center', data = None)
    * x : 막대의 x 좌표
    * height : 막대의 높이
    * width : 막대의 너비
    * bottom : 막대 바닥면의 y 좌표 (누적 막대그래프를 그릴 때 사용)
    * align : x 좌표에 대한 막대 정렬 ('center', 'edge')

```python
# 데이터 가져오기
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
```

```python
wine_load = load_wine()
wine_load
```

```python
wine = pd.DataFrame(wine_load.data, columns=wine_load.feature_names)
wine
```

```python
wine['Class'] = wine_load.target
wine
```

```python
wine['Class'] = wine['Class'].map({0:'class_0', 1:'class_1', 2:'class_2'})
wine
```

```python
wine_type = wine['Class'].value_counts()
wine_type
```

```python
plt.bar(wine_type.index, wine_type.values,               # 데이터의 x, 높이
        width = 0.8, bottom = None,        # 막대의 넓이와 막대 밑면의 y좌표
        align = 'edge')        # 막대의 정렬
plt.show()
```

```python
plt.barh(wine_type.index, wine_type.values, height = 0.8, left = None, align = 'edge')
plt.show()
```
