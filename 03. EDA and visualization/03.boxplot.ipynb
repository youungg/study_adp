# boxplot

### 상자
* 중앙 50% 값
* 상자안 중앙값의 위치에 따라 값이 어디로 치우쳐 있는지 알 수 있다

### IQR
* 중앙 50% 데이터가 퍼진정도
* whisker에서 이상치를 정제하는 판단 기준으로 사용

### whisker
* whisker 에서 벗어난 값은 이상치로 판단
* IQR에 곱해진 상수는 임의로 설정가능



```python
# 데이터 가져오기
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
iris = pd.DataFrame(iris.data, columns=iris.feature_names)
iris
```

```python
iris['Class'] = load_iris().target
iris
```

```python
iris['Class'] = iris['Class'].map({0:'Setosa', 1:'Versicolour', 2:'Virginica'})
iris
```

```python
iris.drop(columns='Class').plot(kind='box')
plt.show()
```

```python
iris.drop(columns ='Class').boxplot(grid = True)
plt.show
```

```python
iris.drop(columns ='Class').boxplot()
plt.show
```

```python
plt.boxplot(iris.drop(columns = 'Class'))
plt.show
```

```python
# whisker 지정 default  = 1.5
plt.boxplot(iris['sepal width (cm)'], whis = 1.5)
plt.title('sepal width (cm)')
plt.show()
```

```python
iris[['sepal width (cm)',  'Class']].boxplot(by = 'Class')
plt.show()
```

```python
sns.boxplot(x="Class", y="sepal width (cm)", data=iris)
plt.show()
```
