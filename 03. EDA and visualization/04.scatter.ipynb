```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = load_iris()
iris
```

```python
iris = pd.DataFrame(iris.data, columns = iris.feature_names)
iris
```

```python
iris['class'] = load_iris().target
iris
```

```python
iris['class'] = iris['class'].map({0:'Setosa', 1: "Versicolor", 2:"Virginica"})
iris
```

```python
plt.title('iris scatter') 
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)') 
```

```python
plt.scatter(x = iris['sepal length (cm)'], y = iris['sepal width (cm)'], alpha = 0.5)
plt.show()
```

```python
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', data=iris, hue='class', style='class')
plt.show()
```
