# 단순선형회귀(simple linear regression model)
* 선형회기 : 입력 특성에 대한 선형함수를 만들어 예측을 하는 알고리즘
* 독립변수가 하나인 경우 특정 직선을 학습하는 것
* 예시

  


* 선형 회기 모델을 이용해 학습데이터 세트에 가장 적합하도록 모델 파라미터를 설정하는 과정을 "학습"
* 회기 성능 평가 지표 : MSE
* 통계적 방식의 회귀분석과 머신러닝은 MSE를 최소화하는 파라미터를 찾는 목표는 같지만, 문제해결 접근방식이 다름
* 통계적 방식의 회귀분석은 정규방정식을 사용해 문제를 해결하지만, 머신러닝 모델은 경사하강법 사용

##  정규방정식


* 위 식을 사용하면 MSE 값을 최소화하는 파라미터 얻기가능
* 최소 자승법(Least Squares Method) 구현 가능

## Linear Regression
* 최소자승법을 사용한 OLS(Ordinary Least Squares) 방식으로 선형 회귀모델을 구현
* 계수 w 를 사용해 선형 모델을 피팅하여 데이터세트에서 관찰된 대상과 선형근사에 의해 예측된 대상간의 잔차제곱합을 최소화
* sklearn.linear_model.LinearRegression (*, fit_intercept = True, normalize = 'deprecated', copy_X = True, n_jobs = None, positive = False)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('C:\\Users\\ljy25\\Desktop\\coding\\adp\\06.supervised learning\\insurance.csv')
data
```

```python
x=data['age']
y=data['charges']
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
x=np.array(data['age'])
y=np.array(data['charges'])
```

```python
print(x.shape)
print(y.shape)
```

```python
# .reshape()으로 차원 맞추기
x=x.reshape(1338 ,1) 
y=y.reshape(1338 ,1)
```

```python
lr = LinearRegression()
lr.fit(x,y)
```

```python
print('선형 회귀 모델 결과')
print('절편', lr.intercept_, '계수', lr.coef_)
```

```python
print(lr.score(x,y))
```

```python
# 19살은 8062원, 64살은 19660원 의료비용을 낸다고 예측
x_new=[[19],[64]]
y_hat=lr.predict(x_new)
print(y_hat)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x_new, y_hat, '-r')
plt.plot(x, y, 'b.')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
