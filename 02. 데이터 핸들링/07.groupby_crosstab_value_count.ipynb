# 데이터 요약

* DataFrame.groupby(by = None, axis = 0, level = None, as_index = True, sort = False, dropna = True).FUN()
    * by : 그룹을 결정하는데 사용
    * axis : 행(0), 열(1) 지정 , default = 0
    * level : 축이 계층 구조일 때, 특정 수준을 기준으로 그룹화
    * as_index : 그룹 레이블이 인덱스로 출력될지의 여부, default = True
    * sort : 집계 행으로 정렬할지 여부
    * dropna : True 이면 결츨값이 행/열 과 함께 삭제, False이면 결측값도 그룹의 키로 처리
</br></br>
* 집계함수 : 여러개의 값을 계산하여 하나의 값으로 만드는 함수
    * count() : 값의 갯수
    * sum() : 값들의 합
    * min() : 최솟값
    * max() : 최댓값
    * mean() : 평균
    * median() : 중앙값
    * std() : 표준편차
    * var() : 분산
    * quantile(n) : 분위수
    * first() : 첫번째 값
    * last() : 마지막 값
    * describe() : 기술 통계량

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

```python
IRIS = load_iris()
IRIS
```

```python
iris = pd.DataFrame(data=IRIS.data, columns=IRIS.feature_names)
iris
```

```python
iris['class'] = IRIS.target
iris['class']
```

```python
iris['class'] = iris['class'].map({0:'setosa', 1:'versicolor', 2:'virginica'})
iris
```

```python
cnt = (iris['class'] == 'setosa').sum()
cnt
```

```python
iris.groupby(by='class').mean()
```

```python
iris.groupby(by='class').median()
```

```python
# value 하나씩 뽑기 -> return counts 하면 개수도 같이 반환
np.unique(iris['petal width (cm)'], return_counts = True)
```

* pd.Series(DataFrame['컬럼명']).value_counts()
    * 수치형 데이터의 경우 구간을 나누지 않고 모든 값에 대해 개수를 센다는 한계가 있음
    * 수치형 데이터를 구간에 따라 나누기 위해 구간을 나눈 컬럼을 생성해야 함

```python
pd.Series(iris['class']).value_counts()
```

```python
iris['petal width level'] = pd.qcut(iris['petal width (cm)'], q=3, labels=['short', 'middle', 'long'])
iris
```

```python
# to_frame() 함수를 사용하여 결과를 데이터프레임으로 변경
pd.Series(iris['petal width level']).value_counts().to_frame()
```

* 두개의 기준에 따른 데이터 분포 확인 방법 :
* pd.crosstab(index, columns, dropna = True, normalize = False)
    * index : 열 위치에서 집계될 범주향 변수
    * columns : 행 위치에서 집계될 범주형 변수
    * dropna : 항목이 모두 NAN인 열을 제외할지 여부
    * normalize : 'all' 혹은 'True' 를 전달하면 모든 값에 대해 정규화 하고,  'index'를 전달하면 각 행에 대해, 'columns'가 전달되면 각 열에 대해 정규화

```python
pd.crosstab(iris['petal width level'], iris['class'])
```
