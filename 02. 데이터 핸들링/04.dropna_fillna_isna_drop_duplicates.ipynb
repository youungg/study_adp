# 조건에 맞는 데이터 탐색 및 수정

```python
import numpy as np
import pandas as pd
```

```python
students = pd.DataFrame({'이름':['장화','홍련','콩쥐','팥쥐','해님','달님'], 
                         '국어': [70, 85, None, 100, None, 85], 
                         '수학':[65, 100, 80, 95, None, 70]})
students
```

# 특정 조건에 해당하는 df 뽑기

- DataFrame[조건식]
    * 조건이 하나일 경우 DataFrame[Dateframe['컬럼명'] >= num]
 </br></br>
- DataFrame [(조건1)& (조건2)& ((조건3)|(조건4))]
    * 조건이 두개 이상일 때

```python
students[students['이름'] == '장화']
```

```python
students[(students['국어'] >=80) & (students['수학'] >=80)]
```

```python
std = students[(students['국어'] >=80) & (students['수학'] >=80)]
std
```

* df.loc[조건문, '추가/변경할 컬럼명']
    * 추가/ 변경할 컬러명이 존재하는 컬럼인 경우 조건문에 부합하는 행의 추가/ 변경할 컬럼명 의 값을 변경
    * 추가/ 변경할 컬럼명이 존재하지 않는 컬럼인 경우, 새로운 컬럼을 만들고 조건문에 부합하는 행에는 지정된 값을 그 외에의 행에는 NAN을 삽입

```python
students.loc[6, '이름':'수학'] = ['별님', 50, 60] # 6라인에 이름부터 수학까지를 넣겠다
students
```

```python
students.loc[(students['국어']>=80)&(students['수학']>=70), '합격'] = 'Pass'
students
```

```python
students.loc[students['합격']!='Pass', '합격'] = 'Fail'
students
```

* np.select(조건목록, 선택목록, default = 디폴트값)
    * 하나의 컬럼에 여러개의 조건에 따라 다른 값을 반환하는 경우 

```python
condition_list = [(students['국어'] >= 90),
                 (students['국어'] >= 80) & (students['국어'] < 90),
                 (students['국어'] >= 70) & (students['국어'] < 80)]
condition_list
```

```python
choice_list = ['A', 'B', 'C']
```

```python
students['점수'] = np.select(condition_list, choice_list, default='F')
students
```

# 결측치 제거
* dropna() 는 default 가 행 계산

* DataFrame.dropna(axis = 0, how = 'any', thresh = None, inplace = False)
    * axis : 0 또는 index 이면 결측값이 포함된 행을 삭제하고 1 또는 columns 이면 결측값이 포함된 열을 삭제
    * how : any 이면 결측값이 존재하는 모든 행/열 삭제, all 이면 모든값이 결측값일 때 삭제
    * thresh : 정수값을 지정하면 결측값이 아닌 값이 그보다 많을 때 행 또는 열을 유지
    * subset : 어떤 레이블에 결측값이 존재하면 삭제할지 정의
    * inplace : True 이면 제자리에서 작업을 수행하고 None 을 반환 -> 원본값 유지가 안됨

```python
students.dropna()
```

```python
# 결측값이 아닌 값이 4보다 많은 행만 남기기 -> 이름도 포함
students.dropna(thresh=4) 
```

```python
students.dropna(thresh=5) 
```

# 결측치 대체
* DataFrame.fillna(value = None, method = None, axis = None, inplace = False, limit = None)
    * value : 단일값 혹은 dict/Series/DataFrame 형식으로 대체할 값을 직접 입력
    * method : 'pad', 'fill' 이전 값으로 채우고, 'backfill', 'bfill' 은 다음에 오는 값으로 채움
    * axis : 0 또는 'index' 이면 행 방향, 1 또는 'columns'이면 열 방향으로 채움
    * limit : method 인자를 지정한 경우 limit으로 지정한 개수만큼만 대체할 수 있음

```python
students.fillna(value=0)
```

```python
health = pd.DataFrame({'연도':[2017, 2018, 2019, 2020, 2021, 2022], 
                       '키': [160, 162, 165, None, None, 166],
                       '몸무게': [53, 52, None, 50, 51, 54],
                       '시력': [1.2, None, 1.2, 1.2, 1.1, 0.8],
                       '병결': [None, None, None, 2, None, 1]})
health
```

```python
health.fillna(0)
```

```python
health.fillna(health.mean())
```

```python
health['병결'] = health['병결'].fillna(0)
health
```

```python
health['몸무게'] = health['몸무게'].fillna(health['몸무게'].mean())
health
```

```python
health.fillna(method='pad', inplace=True)
health
```

# 중복행 삭제
* DataFrame.drop_duplicates()
    * 완전히 중복된 행을 제거하는데 사용
    * 하나의 컬럼에서 구성요소를 확인하거나 몇개의 데이터 세트의 일부분에서 요소들의 조합을 살피는데 사용
    * python 기본 함수인 set 를 대신 사용 가능
    * 이때 set은 출력이 tuple, duplicates는 pd.Series

```python
health['키'].drop_duplicates()
```

```python
health['키'].drop_duplicates().tolist()
```

```python
set(health['키'])
```

```python
health[['시력','병결']].drop_duplicates() # 0,4,5 와 0,3,5 합쳐짐 -> 개별 컬럼에는 중복된 값이 포함될 수 있다
```
