(4) 선택값 변경

```python
score = pd.DataFrame({'국어': [100, 80], '수학':[75, 90], 
                      '영어':[90, 95]}, index=['장화', '홍련'])
score
```

```python
score.loc['홍련', '영어'] = 100
score
```

```python
score['국어'] = score['국어'] - 5
score
```

# row/column 추가

```python
new_students = pd.DataFrame({'국어': [70, 85], '수학':[65, 100], 
                             '영어':[95, 65]}, index=['콩쥐', '팥쥐'])
new_students
```

```python
score = score.append(new_students)
score
```

```python
print(pd.__version__)
```

pandas 2.0.0 버전 이후부터 'append()' Method가 완전히 제거되었기 때문에 더 이상 작동하지 않습니다. 이로 인해 이전에 'append()' Method를 참조하던 코드는 다음과 같은 오류가 발생합니다. 

append를 대체하기 위해 concat을 이용합니다.  -> 이때 대괄호 주의

merge랑 join 도 있음

```python
score1 = pd.concat([score, new_students])
score1
```

```python
new_student1 = pd.Series({'국어':85, '수학':55, '영어':95}, name='해님')
print(new_student1)
score2 = pd.concat([score1,new_student1])
score2
```

```python
new_student1 = pd.Series({'국어':85, '수학':55, '영어':95}, name='해님')
print(new_student1)
score2 = pd.concat([score1,new_student1], axis =1)
score2
```

pd.Series 모양이 달라서 원하는 모양으로 안됨 -> DataFrame으로 변경

```python
new_student1_df = pd.DataFrame([new_student1])
score2 = pd.concat([score, new_students, new_student1_df])
score2
```

ignore_index = True 는 기존 인덱스 사라지는 설정값

```python
new_student2 = {'국어':75, '수학':80, '영어':80}
score3 = pd.concat([score2, new_student2], ignore_index=True)
```

pd.concat 함수는 Series나 DataFrame 객체들만 결합할 수 있습니다. 딕셔너리를 직접 결합하려고 하면 TypeError가 발생합니다.

```python
new_student2_df = pd.DataFrame([new_student2], index=['별님'])  # 새로운 학생의 이름을 인덱스로 지정

# score2와 새로운 학생 데이터 프레임을 결합
score3 = pd.concat([score2, new_student2_df], ignore_index=True)

score3
```
