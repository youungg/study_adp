# 범주형 변수 처리
* 범주형 변수 :
    * 몇 개의 부류나 범위, 서열 등으로 구분하여 수집된 변수로 질적 변수라고도 불림
    * 주로 문자열 데이터로 DataFrame에서 object나 category 형으로 저장
    * 숫자 값인 경우 수학적 의미는 가지지 않기 때문에 덧셈과 같은 수학적 연산의 개념을 적용할 수 없다
</br></br>
* 더미 변수 :
    * 범주형 변수에 있는 범주 각각을 컬럼으로 변경
    * 원본 컬럼의 값이 해당 범주에 속하는 지 여부에 따라 0 or 1 로 채운 변수
    * 수학적 연산이 불가능한 범주형 변수를 True/False 형식의 숫자로 변경하여 수학적 연산을 가능하게 한다. 

* pd.get_dummies(data, columns = ['범주형1', '범주형2'])

```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris = pd.DataFrame(iris.data, columns=iris.feature_names)
iris['Class'] = load_iris().target
iris['Class'] = iris['Class'].map({0:'Setosa', 1:'Versicolour', 2:'Virginica'})
iris
```

```python
iris_dummy = pd.get_dummies(iris, columns = ['Class'])
iris_dummy
```
